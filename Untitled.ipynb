```python
#loading the dataset downloaded from Kaggle
#UCI Machine Learning Repository: https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29
df = pd.read_csv('breast_cancer.csv')
```

```python
# dropping unwanted columns
df.drop(['Unnamed: 32','id'], axis= 1, inplace=True )
```

```python
# Count number of 'zeros' in all columns of Dataframe
for column_name in df.columns:
    column = df[column_name]
    # Get the count of Zeros in column 
    count = (column == 0).sum()
    print('Count of zeros in column ', column_name, ' is : ', count)
```

```python
#checking for duplicate and dropping the same
df.drop_duplicates()
```

```python
#checking number of cases for each category and plotting them
ax = sns.countplot(df.diagnosis,label="Count")       # M = 212, B = 357
B, M = df.diagnosis.value_counts()
print('Number of Benign: ',B)
print('Number of Malignant : ',M)
```

```python
X = df.drop('diagnosis',axis = 1)
y = df.diagnosis
```

```python
# Encoding categorical data from text(B and M) to integers (0 and 1)
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
Y = labelencoder.fit_transform(y) # M = 1 and B = 0
print(np.unique(Y))
```

```python
#correlation map
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(X.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

### FEATURE SELECTION

### LASSO -  has the ability to shrink the not so important variables to 0

```python
#importing the model
from sklearn.linear_model import Lasso

#initiating the model
lasso = Lasso(alpha=0.001)

# fitting on the data
lasso_coef = lasso.fit(X,Y).coef_

#plotting the graph
_=plt.plot(range(len(X.columns)),lasso_coef)
_=plt.xticks(range(len(X.columns)),X.columns,rotation = 90)
_=plt.ylabel('coefficient')
plt.show()
```

```python
lasso_select = []
for i,j in zip(X.columns,lasso_coef):
    if j > 0:
        lasso_select.append(i)
        print (i,j)
```

```python
X = X[lasso_select] 
```

```python
print (X.shape)
X.head()
```

```python
# boxplot to visualize the data variations
X.boxplot()
plt.xticks(range(len(X.columns)),X.columns,rotation = 90)
plt.show()
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    stratify=y, 
                                                    test_size=0.2,random_state=5)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import recall_score,accuracy_score,confusion_matrix,roc_auc_score,roc_curve
```

```python
# Logistic Regression

# importing necessary modules
from sklearn.linear_model import LogisticRegression

# initiate the model
lr_classifier = LogisticRegression(random_state = 51, penalty = 'l2')

# fitting the model
lr_classifier.fit(X_train, Y_train)

#predicting on the test set
y_pred_lr = lr_classifier.predict(X_test)

# scoring
print('Accuracy score : %f' %(accuracy_score(Y_test,y_pred_lr)))
print('Recall score : %f' %((recall_score(Y_test,y_pred_lr))))
print(confusion_matrix(Y_test,y_pred_lr))


fpr, tpr, _ = roc_curve(Y_test, y_pred_lr)
auc = roc_auc_score(Y_test, y_pred_lr)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
# Support vector classifier

# importing necessary modules
from sklearn.svm import SVC

# initiate the model
svc_classifier = SVC()

# fitting the model
svc_classifier.fit(X_train, Y_train)

#predicting on the test set
y_pred_scv = svc_classifier.predict(X_test)
                       

print('Accuracy score : %f' %(accuracy_score(Y_test,y_pred_scv)))
print('Recall score : %f' %((recall_score(Y_test,y_pred_scv))))
print(confusion_matrix(Y_test,y_pred_scv))



fpr, tpr, _ = roc_curve(Y_test, y_pred_scv)
auc = roc_auc_score(Y_test, y_pred_scv)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
# K – Nearest Neighbor Classifier
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn_classifier.fit(X_train, Y_train)
y_pred_knn = knn_classifier.predict(X_test)

print('Accuracy score : %f' %(accuracy_score(Y_test,y_pred_knn)))
print('Recall score : %f' %((recall_score(Y_test,y_pred_knn))))
print(confusion_matrix(Y_test,y_pred_knn))

fpr, tpr, _ = roc_curve(Y_test, y_pred_knn)
auc = roc_auc_score(Y_test, y_pred_knn)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
# Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, Y_train)
y_pred_nb = nb_classifier.predict(X_test)

print('Accuracy score : %f' %(accuracy_score(Y_test,y_pred_nb)))
print('Recall score : %f' %((recall_score(Y_test,y_pred_nb))))
print(confusion_matrix(Y_test,y_pred_scv))

fpr, tpr, _ = roc_curve(Y_test, y_pred_nb)
auc = roc_auc_score(Y_test, y_pred_nb)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier



# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(random_state= 42)  

scoring = {"Recall": 'recall', "Accuracy": 'accuracy'}

cv = StratifiedKFold(n_splits= 10, shuffle=True, random_state=1)
parameter = {
             'n_estimators': [200,300,400,500],
             'max_depth' : [4,5,6,7,8],
             'criterion' :['gini', 'entropy'],
             'min_samples_leaf': [3, 4, 5],
             'min_samples_split': [8, 10, 12],
             'bootstrap': [True, False] # boot straping method is used to build the tree rather than using whole dataset

            }

rf_classifier = GridSearchCV(estimator = model_rf, 
                         param_grid = parameter, 
                         cv = cv, 
                         scoring= scoring,
                         refit="Recall", #refits with the parameter setting that has the best cross-validated score.
                         n_jobs = -1, #means using all processors
                         return_train_score =False # to get insights on how different parameter settings impact the overfitting/underfitting trade-off.
                        )

rf_classifier.fit(X_train, Y_train)
print(rf_classifier.best_params_)
print(rf_classifier.best_score_)

y_pred_rf = rf_classifier.predict(X_test)

print('Accuracy score : %f' %(accuracy_score(Y_test,y_pred_rf)))
print('Recall score : %f' %((recall_score(Y_test,y_pred_rf))))
print(confusion_matrix(Y_test,y_pred_rf))

fpr, tpr, _ = roc_curve(Y_test, y_pred_rf)
auc = roc_auc_score(Y_test, y_pred_rf)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
# Randomized Search
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score,accuracy_score,recall_score

params={
 "learning_rate"    : [ 0.10, 0.30 ] ,
 "max_depth"        : [ 3, 4, 6],
 "min_child_weight" : [ 3, 5, 7 ],
 "gamma"            : [ 0.2, 0.3, 0.4 ],
 "alpha"            : [0.1,0.5,1,5],
 "colsample_bytree" : [ 0.3, 0.4, 0.7 ] 
}

model_xgb = XGBClassifier()


xgb_classifier = GridSearchCV(model_xgb, param_grid=params, cv = 5,scoring= 'roc_auc', n_jobs= -1, verbose= 3)

xgb_classifier.fit(X_train, Y_train)

print (xgb_classifier.best_estimator_)

y_pred_xgb = xgb_classifier.predict(X_test)

print('Accuracy score : %f' %(accuracy_score(Y_test,y_pred_xgb)))
print('Recall score : %f' %((recall_score(Y_test,y_pred_xgb))))
print(confusion_matrix(Y_test,y_pred_xgb))

fpr, tpr, _ = roc_curve(Y_test, y_pred_xgb)
auc = roc_auc_score(Y_test, y_pred_xgb)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
from catboost import CatBoostClassifier

params = {'iterations':5000,
        'learning_rate':0.001,
        'depth':3,
        'eval_metric':'AUC',
        'verbose':200,
        'od_type':"Iter", # overfit detector
        'od_wait':500, # most recent best iteration to wait before stopping
        'random_seed': 1
          }

cat_model = CatBoostClassifier(**params)
cat_model.fit(X_train, Y_train,   
          eval_set=(X_test, Y_test), 
          use_best_model=True, # True if we don't want to save trees created after iteration with the best validation score
          plot=True  
         );
y_pred_cat = cat_model.predict(X_test)


print('Accuracy score : %f' %(accuracy_score(Y_test,y_pred_cat)))
print('Recall score : %f' %((recall_score(Y_test,y_pred_cat))))
print(confusion_matrix(Y_test,y_pred_cat))

fpr, tpr, _ = roc_curve(Y_test, y_pred_cat)
auc = roc_auc_score(Y_test, y_pred_cat)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
